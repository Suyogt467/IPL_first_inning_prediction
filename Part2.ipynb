merge_df['batting_team'] = np.where(merge_df['toss_decision'] == 'bat', merge_df['team1'], merge_df['team2'])
merge_df['bowling_team'] = np.where(merge_df['toss_decision'] == 'bat', merge_df['team2'], merge_df['team1'])

def batting(row):
    if row['toss_won'] == row['team1'] and row['toss_decision'] == 'bat':
        return row['team1']
    elif row['toss_won'] == row['team2'] and row['toss_decision'] == 'bat':
        return row['team2']
    else:
        return ''

def field(row):
    if row['toss_won'] == row['team1'] and row['toss_decision'] == 'field':
        return row['team1']
    elif row['toss_won'] == row['team2'] and row['toss_decision'] == 'field':
        return row['team2']
    else:
        return ''

merge_df['batting_team'] = merge_df.apply(batting, axis=1)
merge_df['bowling_team'] = merge_df.apply(field, axis=1)

```python
t1_list = []
t2_list = []

# Iterate through each row in the DataFrame
for index, row in merge_df.iterrows():
    toss_won_team = row['toss_won']
    teams_list = row['teams'].split(',')  # Convert the string to a list
    
    # Check if toss_won_team is in the teams_list
    if toss_won_team in teams_list:
        t1_list.append(toss_won_team)
        # Remove the toss_won_team from the list to get the remaining teams
        teams_list.remove(toss_won_team)
        t2_list.append(teams_list)
    else:
        t1_list.append(None)  # If t2 has the value, t1 should be None
        t2_list.append(teams_list)

# Add the t1 and t2 lists as new columns in the DataFrame
merge_df['t1'] = t1_list
merge_df['t2'] = t2_list
```

```python
for index, row in merge_df.iterrows():
    if row['toss_decision'] == 'bat':
        merge_df.at[index, 'batting'] = row['t1']
        merge_df.at[index, 'bowling'] = None
    elif row['toss_decision'] == 'field':
        merge_df.at[index, 'batting'] = None
        merge_df.at[index, 'bowling'] = row['t1']

    # Apply the conditions for t2 batting and bowling
    if row['toss_decision'] == 'bat':
        merge_df.at[index, 'bowling'] = row['t2']
        merge_df.at[index, 'batting'] = None
    elif row['toss_decision'] == 'field':
        merge_df.at[index, 'bowling'] = None
        merge_df.at[index, 'batting'] = row['t2']
```

```python
merge_df['batting'] = merge_df.apply(lambda row: row['t1'] if row['toss_decision'] == 'bat' else row['t2'], axis=1)
merge_df['bowling'] = merge_df.apply(lambda row: row['t2'] if row['toss_decision'] == 'bat' else row['t1'], axis=1)
```

```python
merge_df['over_no'] = merge_df['over'].apply(lambda x: int(x) if isinstance(x, int) else int(str(x).split(".")[0]))

def get_ball_no(x):
    if isinstance(x, int):
        return 0
    else:
        split_result = str(x).split(".")
        if len(split_result) > 1:
            return int(split_result[1])
        else:
            return 0  # Default value for ball number if dot is not found

merge_df['ball_no'] = merge_df['over'].apply(get_ball_no)
```

```python
merge_df['ball_count'] = (merge_df['over'].astype('int')*6) + merge_df['ball_no'].astype('int')
```

```python
merge_df
```

```python
merge_df['current_score'] = merge_df.groupby('match_no').cumsum()['score']
```

```python
merge_df['current_score'].unique()
```

```python
merge_df['current_runrate'] = (merge_df['current_score']/merge_df['ball_count'])*6
```

```python
merge_df['player_dismissed'] = merge_df['outcome'].apply(lambda x:1 if x=='w' else 0)
merge_df['player_dismissed'] = merge_df['player_dismissed'].astype(int)
merge_df['player_dismissed'] = merge_df.groupby('match_no').cumsum()['player_dismissed']
merge_df['wickets_left'] = 10 - merge_df['player_dismissed']
```

```python
# Last five overs runrate
groups = merge_df.groupby('match_no')

match_ids = merge_df['match_no'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['score'].values.tolist())
```

```python
merge_df['last_five'] = last_five
```
