```python
# Bowling team
merged_df['Teams'] = merged_df.apply(lambda row: row['team2'] + ',' + row['team1'], axis=1)
```

```python
def get_bowling_team(row):
    teams = row['Teams'].split(',')
    return teams[0] if teams[0] != row['BattingTeam'] else teams[1]
```

```python
merged_df['BowlingTeam'] = merged_df.apply(get_bowling_team, axis=1)
```

```python
merged_df.drop(columns=['Teams'],inplace=True)
```

```python

merged_df['player_dismissed'] = merged_df.groupby('ID').cumsum()['isWicketDelivery']
merged_df['wickets_left'] = 10 - merged_df['player_dismissed']
```

```python
merged_df.drop(columns=['isWicketDelivery','player_out'],inplace =True)
```

```python
merged_df['ball_bowled'] = merged_df['extra_type'].apply(lambda x: 0 if x in ['noballs', 'wides'] else 1)
```

```python
merged_df['ball_bowled']
```

```python
merged_df['ball_bowled'].value_counts()
```

```python
merged_df['total_ball_bowled'] = merged_df.groupby('ID')['ball_bowled'].cumsum()
```

```python
merged_df['total_ball_bowled'] = merged_df['total_ball_bowled']
```

```python
merged_df['balls_left'] = 120 - merged_df['total_ball_bowled']
merged_df['balls_left'] = merged_df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
merged_df
```

```python
merged_df['current_score'] = merged_df.groupby('ID')['total_run'].cumsum() 
```

```python
merged_df
```

```python
merged_df['current_runrate'] = merged_df['current_score']/merged_df['total_ball_bowled']*6
```

```python
merged_df
```

```python
# Last five over runrate
groups = merged_df.groupby('ID')

match_ids = merged_df['ID'].unique()
last_five_over_runs = []
for id in match_ids:
    last_five_over_runs.extend(groups.get_group(id).rolling(window=30).sum()['total_run'].values.tolist())
    
```

```python
merged_df['last_five_over_runs'] = last_five_over_runs
```
